# 模型預測

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam


tokenizer = Tokenizer()

tokenizer.fit_on_texts(sent_all)

#定義總字數。有用來填充的0 所以+1。
total_words = len(tokenizer.word_index) + 1

print(f'word index dictionary: {tokenizer.word_index}')
print(f'total words: {total_words}')
```

```python
month = 12
day = 31
spoken = 14
preiod = 4


sequences_month_train = tokenizer.texts_to_sequences(sent_month_train)
sequences_month_test = tokenizer.texts_to_sequences(sent_month_test)
sequences_day_train = tokenizer.texts_to_sequences(sent_day_train)
sequences_day_test = tokenizer.texts_to_sequences(sent_day_test)
sequences_spoken_train = tokenizer.texts_to_sequences(sent_spoken_train)
sequences_spoken_test = tokenizer.texts_to_sequences(sent_spoken_test)
sequences_preiod_train = tokenizer.texts_to_sequences(sent_preiod_train)
sequences_preiod_test = tokenizer.texts_to_sequences(sent_preiod_test)


max_sequence_month_train = max([len(x) for x in sequences_month_train])
max_sequence_month_test = max([len(x) for x in sequences_month_test])
max_sequence_day_train = max([len(x) for x in sequences_day_train])
max_sequence_day_test = max([len(x) for x in sequences_day_test])
max_sequence_spoken_train = max([len(x) for x in sequences_spoken_train])
max_sequence_spoken_test = max([len(x) for x in sequences_spoken_test])
max_sequence_preiod_train = max([len(x) for x in sequences_preiod_train])
max_sequence_preiod_test = max([len(x) for x in sequences_preiod_test])

max_sequence_month = max(max_sequence_month_train ,max_sequence_month_test)
max_sequence_day = max(max_sequence_day_train, max_sequence_day_test)
max_sequence_spoken = max(max_sequence_spoken_train, max_sequence_spoken_test)
max_sequence_preiod = max(max_sequence_preiod_train, max_sequence_preiod_test)

input_sequences_month_train = np.array(pad_sequences(sequences_month_train, maxlen=max_sequence_month, padding='pre'))
input_sequences_month_test = np.array(pad_sequences(sequences_month_test, maxlen=max_sequence_month, padding='pre'))
input_sequences_day_train = np.array(pad_sequences(sequences_day_train, maxlen=max_sequence_day, padding='pre'))
input_sequences_day_test = np.array(pad_sequences(sequences_day_test, maxlen=max_sequence_day, padding='pre'))
input_sequences_spoken_train = np.array(pad_sequences(sequences_spoken_train, maxlen=max_sequence_spoken, padding='pre'))
input_sequences_spoken_test = np.array(pad_sequences(sequences_spoken_test, maxlen=max_sequence_spoken, padding='pre'))
input_sequences_preiod_train = np.array(pad_sequences(sequences_preiod_train, maxlen=max_sequence_preiod, padding='pre'))
input_sequences_preiod_test = np.array(pad_sequences(sequences_preiod_test, maxlen=max_sequence_preiod, padding='pre'))


ys_month_train = tf.keras.utils.to_categorical(label_month_train, num_classes= month+1)
ys_month_test = tf.keras.utils.to_categorical(label_month_test, num_classes= month+1)

ys_day_train = tf.keras.utils.to_categorical(label_day_train, num_classes= day+1)
ys_day_test = tf.keras.utils.to_categorical(label_day_test, num_classes= day+1)

ys_spoken_train = tf.keras.utils.to_categorical(label_spoken_train, num_classes= spoken+1)
ys_spoken_test = tf.keras.utils.to_categorical(label_spoken_test, num_classes= spoken+1)

ys_preiod_train = tf.keras.utils.to_categorical(label_preiod_train, num_classes= preiod+1)
ys_preiod_test = tf.keras.utils.to_categorical(label_preiod_test, num_classes= preiod+1)
```

```python
#month

import tensorflow as tf

# Define dimensionality of the embedding
embedding_dim = 64

# Build the model
model_month = tf.keras.Sequential([
    tf.keras.layers.Embedding(total_words, embedding_dim, input_length=max_sequence_month),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(13, activation='softmax')
])

# Print the model summary
model_month.summary()
```

```python
num_epochs = 10

# Set the training parameters
model_month.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

# Start training
history = model_month.fit(input_sequences_month_train ,ys_month_train ,epochs=num_epochs , validation_data=(input_sequences_month_test, ys_month_test))
```

```python
#day

import tensorflow as tf

# Define dimensionality of the embedding
embedding_dim = 64

# Build the model
model_day = tf.keras.Sequential([
    tf.keras.layers.Embedding(total_words, embedding_dim, input_length=max_sequence_day),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(32, activation='softmax')
])

# Print the model summary
model_day.summary()
```

```python

def model_predict(seed_text):
  error_ = False
  error_spoken = False

  line = seed_text

  line = line.strip('\n')   #將字串前後的 換行處理掉
  line = line.replace(" ", "")  #將字串中空格刪除
  line = line.replace("？", "")  
  line = line.replace("，", "")  
  line = line.replace("。", "")  
  line = line.replace("?", "")  
  line = line.replace("！", "")  
  line = line.replace("...", "")  
  seed_text = line

  seed_text = str(HanLP.segment(HanLP.tw2s(seed_text)))

  print(seed_text)

  # 定義要預測的總詞數

  token_list = tokenizer.texts_to_sequences([seed_text])[0]
  print("序列:",token_list)
  token_list_month = pad_sequences([token_list], maxlen=max_sequence_month, padding='pre')
  token_list_day = pad_sequences([token_list], maxlen=max_sequence_day, padding='pre')
  print("填充(月份):",token_list_month)
  print("填充(日期):",token_list_day)
  output_month = model_month.predict(token_list_month)
  output_day = model_day.predict(token_list_day)
  predicted_month = np.argmax(output_month)
  predicted_day = np.argmax(output_day)

  if predicted_month != 0 and predicted_day != 0:
    print("預測月份:",predicted_month)
    print("預測日期:",predicted_day)
  else: 
    error_ = True


  #口語

  token_list = tokenizer.texts_to_sequences([seed_text])[0]
  print("序列:",token_list)
  token_list = pad_sequences([token_list], maxlen=max_sequence_spoken, padding='pre')
  print("填充:",token_list)
  output_spoken = model_spoken.predict(token_list)

  spoken_2 = {1:"禮拜一" , 2:"禮拜二" , 3:"禮拜三" , 4:"禮拜四" , 5:"禮拜五" , 6:"禮拜六" , 7:"禮拜日" ,8:"下禮拜一" , 9:"下禮拜二" , 10:"下禮拜三" , 11:"下禮拜四" , 12:"下禮拜五" , 13:"下禮拜六" , 14:"下禮拜日" }

  predicted_spoken = np.argmax(output_spoken)
  if predicted_spoken != 0 :
    print("預測數字",predicted_spoken)
    print("預測轉換:",spoken_2[predicted_spoken])
  else:
    error_spoken = True


  token_list = tokenizer.texts_to_sequences([seed_text])[0]
  print("序列:",token_list)
  token_list = pad_sequences([token_list], maxlen=max_sequence_preiod, padding='pre')
  print("填充:",token_list)
  output_preiod = model_preiod.predict(token_list)

  preiod_2 = {1:"早上" , 2:"中午" , 3:"晚上" , 4:"凌晨" }

  predicted_preiod = np.argmax(output_preiod)
  if predicted_preiod != 0 :
    print("預測數字",predicted_preiod)
    print("預測轉換:",preiod_2[predicted_preiod])
  else:
    error_preiod = True



  if error_ == True and error_spoken == True and error_preiod == True:
    print("錯誤")
```
